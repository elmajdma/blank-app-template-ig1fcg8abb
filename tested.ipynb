```python
import altair as alt
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit

# Generate some example data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = 3 * x**2 + 2 * x + 1 + np.random.normal(size=x.size)

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Define a polynomial function for curve fitting
def poly2(x, a, b, c):
    return a * x**2 + b * x + c

# Perform curve fitting
params, _ = curve_fit(poly2, data['x'], data['y'])

# Generate fitted values
data['y_fit'] = poly2(data['x'], *params)

# Create the scatter plot and the fitted line plot
scatter = alt.Chart(data).mark_circle().encode(
    x='x',
    y='y'
)

line = alt.Chart(data).mark_line(color='red').encode(
    x='x',
    y='y_fit'
)

# Combine the plots
chart = scatter + line
chart.show()
```
